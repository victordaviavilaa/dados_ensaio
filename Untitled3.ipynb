```python
import csv

dados = [
["Ensaio_ID", "Forca_N", "Deslocamento_mm"],
["E-001", "2450.0", "1.2"],
["E-002", "3800.0", "1.8"],
["E-003", "4500.0", "2.1"],
["E-004", "1200.0", "0.6"],
["E-005", "5100.0", "3.0"]
]

with open("dados_ensaio.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(dados)

print("CSV criado")

ids = []
forca = []
desloc = []

with open("dados_ensaio.csv", "r") as f:
    linhas = f.readlines()

for linha in linhas[1:]:
    partes = linha.strip().split(",")

    ids.append(partes[0])
    forca.append(float(partes[1]))
    desloc.append(float(partes[2]))

print(ids)


import math

def tensao(f, diametro=8):
    area = math.pi * (diametro/2)**2
    return f / area

def deformacao(x, l0=50):
    return x / l0





tensoes = []
deformacoes = []
classes = []

for i in range(len(ids)):
    t = tensao(forca[i])
    e = deformacao(desloc[i])

    tensoes.append(t)
    deformacoes.append(e)

    if t > 80:
        classes.append("Classe A")
    elif t >= 40:
        classes.append("Classe B")
    else:
        classes.append("Classe C")

print(tensoes)
print(classes)


with open("relatorio.txt", "w") as r:
    r.write("=== SCPET-10 RELATORIO ===\n\n")

    for i in range(len(ids)):
        r.write(f"{ids[i]} | {tensoes[i]:.2f} MPa | {classes[i]}\n")

print("Relatório gerado")
print(forca)
print(desloc)
```
